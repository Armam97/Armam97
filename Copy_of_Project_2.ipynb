```python
ax=sns.countplot(x='Gender',data=df)
for bar in ax.containers :
     ax.bar_label(bar)
```

```python
ax=sns.countplot(data=df,x='Gender',hue='Ever_Married')
for bar in ax.containers :
     ax.bar_label(bar)
```

```python
ax=sns.countplot(data=df,x='Profession',hue='Spending_Score')
for bar in ax.containers :
     ax.bar_label(bar)
```

```python
ax=sns.countplot(data=df,x='Profession',hue='Gender')
for bar in ax.containers :
     ax.bar_label(bar)
```

```python
ax=sns.countplot(data=df,x='Spending_Score',hue='Ever_Married')
for bar in ax.containers :
     ax.bar_label(bar)
```

```python
grouped_data = df.groupby(['Spending_Score', 'Gender'], as_index=False).size()
# Create a bar chart for Gender and Spending Score
sns.set(rc={'figure.figsize': (10, 6)})
sns.barplot(data=grouped_data, x='Gender', y='size', hue='Spending_Score')
plt.title('Spending Score vs Gender')
plt.show()
```
